<strong>Simple Tic Tac Toe Game</strong>

Instructions - Use NumPad to position your marker(X/O)

```python
def display_board(board):
    
    print('\n')
    print(board[7] + '|' + board[8] + '|' + board[9])
    print('-----')
    print(board[4] + '|' + board[5] + '|' + board[6])
    print('-----')
    print(board[1] + '|' + board[2] + '|' + board[3])
    print('\n')
        
```

```python
def player_input():
    
    marker = ''
    
    while(marker.upper() != 'X' and marker.upper() != 'O'):
        marker = input('\nPlayer1, Please pick a marker (X/O): ')
    
    player1 = marker
    
    if player1.upper() == 'X':
        player2 = 'O'
    else:
        player2 = 'X'
        
    return (player1, player2)
        
```

```python
def place_marker(board, marker, position):
    
    board[position] = marker
```

```python
def win_check(board, mark):
    
    if board[1] == mark and board[2] == mark and board[3] == mark:
        return True
    elif board[4] == mark and board[5] == mark and board[6] == mark:
        return True
    elif board[7] == mark and board[8] == mark and board[9] == mark:
        return True
    elif board[1] == mark and board[4] == mark and board[7] == mark:
        return True
    elif board[2] == mark and board[5] == mark and board[8] == mark:
        return True
    elif board[3] == mark and board[6] == mark and board[9] == mark:
        return True
    elif board[1] == mark and board[5] == mark and board[9] == mark:
        return True
    elif board[3] == mark and board[5] == mark and board[7] == mark:
        return True
    else:
        return False
```

```python
import random

def choose_first():
    
    if random.randint(1, 2) == 1:
        return 'Player1'
    else:
        return 'Player2'
```

```python
def space_check(board, position):
    
    return board[position] == ' '
```

```python
def full_board_check(board):
    
    for item in board:
        if item == ' ':
            return False
    else:
        return True
```

```python
def player_choice(board):
    
    nextPosition = 0
    
    while(nextPosition not in range(1, 10)):
        nextPosition = int(input('Enter your position (1-9): '))
        
    if space_check(board, nextPosition):
        return nextPosition
```

```python
def replay():
    
    replay = input('Do you want to play again? (Y/N): ')
  
    return replay.lower() == 'y'
```

```python
# Gameplay

print('\nWelcome to Tic Tac Toe!\n')

while True:
    
    player1_marker, player2_marker = player_input()
    
    firstPlayer = choose_first()
    
    if firstPlayer == 'Player1':
        secondPlayer = 'Player2'
        firstPlayer_marker = player1_marker
        secondPlayer_marker = player2_marker
    else:
        secondPlayer = 'Player1'
        firstPlayer_marker = player2_marker
        secondPlayer_marker = player1_marker
        
    print('\n' + firstPlayer + ' will go first!')
    
    board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']

    count = 1
    
    while True:
            
        display_board(board)
        
        if count%2 != 0:
            print(firstPlayer + "'s turn")
            
            nextPos = player_choice(board)
            
            if not nextPos:
                continue
                
            place_marker(board, firstPlayer_marker, nextPos)
            
            if win_check(board, firstPlayer_marker):
                print('\n' + firstPlayer + ' has won the game!\n')
                display_board(board)
                break
        else:
            print(secondPlayer + "'s turn")
            
            nextPos = player_choice(board)
            
            if not nextPos:
                continue
                
            place_marker(board, secondPlayer_marker, nextPos)
            
            if win_check(board, secondPlayer_marker):
                print('\n' + secondPlayer + ' has won the game!')
                display_board(board)
                break
            
            
        if full_board_check(board):
            print("\nIt's a draw!\n")
            display_board(board)
            break
            
        count += 1

    if not replay():
        break
```
